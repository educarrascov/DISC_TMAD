El archivo `DatosT1.csv` contiene datos georeferenciados de los accidentes de tránsito en que involucraron a peatones en la ciudad de Santiago durante 2015. Cada fila contiene un evento, y las columnas son las siguientes:
* X:  longitud geografica del evento
* Y:  latitud geografica del evento
* Ubicacion: dirección en la que fue el evento
* Fallecidos: número de fallecidos involucrados.
* Graves: número de personas con heridas graves.
* MenosGrave: número de personas cor heridas menos graves.
* Leve: número de personas con heridas leves.
* Accidentes: número de accidentes en el mismo lugar.

## Actividad 1: 

Considere que las variables `X0` e `Y0` representan las coordenadas de un punto en el mapa. Por ejemplo:

````python
X0 = -70.51986870520346
Y0 = -33.48854784792859
````

Corresponden a la entrada del campus Peñalolén.

Escriba un programa que encuentre el evento más cercano al punto dado por `(X0,Y0)` e imprima la información asociada al evento. Su programa debe entrega la información  en el formato que se muestra a continuación:


```
El accidente más cercano a la ubicación idicada ocurrió en DIAGONAL LAS TORRES & CAMINO PRIVADO; PENALOLEN y tuvo 0 fallecidos, 1 heridos graves, 0 heridos menos graves y 0 heridos leves.
```

**Observación:** El punto (X0,Y0) anterior se entrega solo a modo de ejemplo. Para la corrección se podría considerar otro punto (X0,Y0). Su programa debe funcionar para cualquier punto definido en el bloque anterior. 

````python
# Escriba aquí su solución.

# Entrada de datos

X0 = -70.51986870520346
Y0 = -33.48854784792859

# Programa

# Abrir, recorrer y cerrar archivo csv
archivo = open("datosT1.csv", "r")  # Abrir archivo CSV 
lista_archivo = archivo.readlines() # Recorrer todo el archivo CSV y devuelve listas de strings.
# Al usar este comando, se guarda el archivo completo, y por tanto no es necesario usar comando close()

# Dado que hay que calular una distancia hacia un punto,
# entonces es necesario establecer una distancia de referencia para posteriormente comparar y
# poder determinar la distancia minima. Este valor puede ser muy pequeño o muy grande, en este caso se eligio el infinito
dist_ref = float("inf")

# Se crea una lista vacía para guardar datos de los calculos de distancia. Se le denomino filas ya que eso va a ir recorriendo.
fila = []

# Procede a recorrer el archivo mediante un ciclo que recorra las filas, calcule la distancia, compare (un condicional) las distancias y guarde los datos en la lista vacía.

for i in range(1, len(lista_archivo)):  # Recorrer desde una variable inicial i por todo el largo del archivo.
    linea = lista_archivo[i].split(",") # Usamos el metodo split para saltar cada coma del archivo 
    
    # Calulamos la distancia utilizando la ecuación de distancia euclediana entre 2 puntos
    
    dist_euclediana = (((float(linea[0])-X0) ** 2) + ((float(linea[1])-Y0) ** 2) ** (0.5))
    
    # Si la distancia referencial es mas grande que la calculada entonces guarda como referencial la nueva distancia calulada e ir guardando en la lista vacia.
    if dist_ref > dist_euclediana:
        dist_ref = dist_euclediana
        fila = linea

# Salida de datos

print("El accidente más cercano a la ubicación indicada ocurrió en", fila[2].replace('"',""), # Es necesario usar replace para eliminar las comillas de la ubicación y salga en el formato solicitado
      "y tuvo", fila[3],
      "fallecidos,", fila[4],
     "heridos graves,", fila[5],
     "heridos menos graves y", fila[6],
     "heridos leves.")
````

## Actividad 2

Escriba un programa que solicite al usuario ingresar el nombre de una comuna (en mayúsculas y sin tildes) e imprima un histograma con la información asociada al total de accidentes ocurridos en esa comuna durante 2015. Además, para cada ítem, entregue los porcentajes respecto a todas las comunas. 

Su solución debe tener el formato que se entrega en el siguiente ejemplo. 

**Ejemplo**
```
Ingrese el nombre de una comuna: PENALOLEN

Histograma de accidentes durante el 2015 para la comuna de PENALOLEN:

Número de accidentes:            68  ->  3.10%
Número de fallecidos:             3  ->  2.83%
Número de heridos graves:        19  ->  3.49%
Número de heridos menos graves:   7  ->  3.87%
Número de heridos leves:         59  ->  3.87%`
```

**Observación:** Si el usuario indica un valor para el que no se tienen datos, su programa debe imprimir un mensaje 
```
No se tiene datos para esa comuna
```

````python
# Escriba aquí su solución.

# Abre el archivo CSV en modo lectura
archivo = open("datosT1.csv", "r")  # Abrir archivo CSV 
lineas = archivo.readlines()

# Solicita el nombre de la comuna
comuna = input("Ingrese el nombre de una comuna: ")

# Inicializa los acumuladores a cero
total_accidentes = 0
total_fallecidos = 0
total_graves = 0
total_menos_graves = 0
total_leves = 0

accidentes_comuna = 0
fallecidos_comuna = 0
graves_comuna = 0
menos_graves_comuna = 0
leves_comuna = 0

# Itera sobre las líneas del archivo CSV
for linea in lineas[1:]:
    # Separa la línea en una lista de valores
    valores = linea.strip().split(',')
    
    # Extrae el nombre de la comuna de la lista de valores
    nombre_comuna = valores[2].split(";")[1].strip('" ')

    # Actualiza los acumuladores totales
    total_accidentes = total_accidentes + int(valores[7])
    total_fallecidos = total_fallecidos + int(valores[3])
    total_graves = total_graves + int(valores[4])
    total_menos_graves = total_menos_graves + int(valores[5])
    total_leves = total_leves + int(valores[6])
    
    # Verifica si la comuna actual coincide con la comuna deseada
    if nombre_comuna == comuna:
        # Actualiza los acumuladores para la comuna deseada
        accidentes_comuna = accidentes_comuna + int(valores[7])
        fallecidos_comuna = fallecidos_comuna + int(valores[3])
        graves_comuna = graves_comuna + int(valores[4])
        menos_graves_comuna = menos_graves_comuna + int(valores[5])
        leves_comuna = leves_comuna + int(valores[6])

        
# Calcula los porcentajes para la comuna deseada
porcentaje_accidentes =  (accidentes_comuna / total_accidentes) * 100
porcentaje_fallecidos =  (fallecidos_comuna / total_fallecidos) * 100
porcentaje_graves =  (graves_comuna / total_graves) * 100
porcentaje_menos_graves =  (menos_graves_comuna / total_menos_graves) * 100
porcentaje_leves = (leves_comuna / total_leves) * 100
    
# Verifica si se encontraron datos para la comuna ingresada en la variable input
if accidentes_comuna == 0:
    print("No se tiene datos para esa comuna")
else:
# Imprime los resultados segun formato solicitado
    print(f" ")
    print(f"Histograma de accidentes durante el 2015 para la comuna de {comuna}:")
    print(f" ")
    print(f"Número de accidentes:          {accidentes_comuna:5d} -> {porcentaje_accidentes:6.2f}%")
    print(f"Número de fallecidos:          {fallecidos_comuna:5d} -> {porcentaje_fallecidos:6.2f}%")
    print(f"Número de heridos graves:      {graves_comuna:5d} -> {porcentaje_graves:6.2f}%")
    print(f"Número de heridos menos graves:{menos_graves_comuna:5d} -> {porcentaje_menos_graves:6.2f}%")
    print(f"Número de heridos leves:       {leves_comuna:5d} -> {porcentaje_leves:6.2f}%")
````

````python
# Escriba aquí su solución.

# Lectura de archivo
archivo = open('datosT1.csv', 'r')
next(archivo)  # Saltar encabezado

# Listas vacias para guardar datos de las comunas y de toda la información 

data = []
comunas = []

# Recorrer las lineas del archivo con los arreglos de las comunas

for linea in archivo:
    lineas = linea.strip().split(',')
    comuna = lineas[2].split('; ')[-1].replace('"', '') # Obtener el nombre de la comuna mediante comando split desde ";" ya que allí esta la comuna y se usa replace para eliminar comillas.
    accidentes = int(lineas[7])
    fallecidos = int(lineas[3])
    graves = int(lineas[4])
    menos_graves = int(lineas[5])
    leves = int(lineas[6])
        
# Si esta la comuna previamente repetida no se guarda ya que esta repetida y si no continuar con la suma acumulada de los datos

    if comuna not in comunas: 
        comunas.append(comuna) # Ir guardando comunas con función append en lista comunas
        data.append([comuna, accidentes, fallecidos, graves, menos_graves, leves]) # Ir guardando el resto de variables con función append en lista data
    else:
        for total in data:
            if total[0] == comuna:
                total[1] = total[1] + accidentes
                total[2] = total[2] + fallecidos
                total[3] = total[3] + graves
                total[4] = total[4] + menos_graves
                total[5] = total[5] + leves
                

# Imprimir tabla segun la estructura solicitada

print('|{:<21}|{:^12}|{:^12}|{:^8}|{:^15}|{:^8}|'.format('Comuna', 'Accidentes', 'Fallecidos', 'Graves', 'Menos Graves', 'Leves'))
print('-' * 83)
for total in data:
    print('|{:>21}|{:>12}|{:>12}|{:>8}|{:>15}|{:>8}|'.format(total[0], total[1], total[2], total[3], total[4], total[5]))
print('-' * 83)

# Imprimir totales

accidentes = sum(total[1] for total in data)
fallecidos = sum(total[2] for total in data)
graves = sum(total[3] for total in data)
menos_graves = sum(total[4] for total in data)
leves = sum(total[5] for total in data)
print('|{:>21}|{:>12}|{:>12}|{:>8}|{:>15}|{:>8}|'.format('TOTAL', accidentes, fallecidos, graves, menos_graves, leves))

# Cerrar archivo
archivo.close()
````
