```python
import pandas as pd
import numpy as np
def getInfo (archivo, columna):
    """archivo en csv y nombre de columna"""
    data_frame = pd.read_csv(archivo, header=0, sep=",", encoding="utf-8")
    if columna not in data_frame.columns:
        return "Lista Vacía"
    ## Generar descripción de columna
    datos_columna = data_frame[columna].value_counts()
  
    ## Verificar si es numérica
    lista = datos_columna.tolist()
  
    ## Si es numérica, convertir a float 
    for estado in lista:
        try:
            float(estado)
        except ValueError:
            return lista
## Si no es numérica, devolver lista
    return [float(i) for i in lista] 
```

```python
def makeHistogram(lista):
    """
    En caso valores numéricos, estadísticas básicas.
    En caso valores no numéricos (categóricos), histograma frecuencia.
    """
    if not lista:
        return "Lista Vacía"
    
    # Comprobamos si la lista contiene valores numéricos o no
    try:
        valores = np.array(lista, dtype=float)
        
        # Calculamos las estadísticas básicas
        minimo = np.min(valores)
        promedio = np.mean(valores)
        desv_estandar = np.std(valores)
        maximo = np.max(valores)
        
        # Mostramos las estadísticas básicas en pantalla
        print(f"Mínimo de la lista: {minimo:.2f}")
        print(f"Promedio de la lista: {promedio:.2f}")
        print(f"Desviación estándar de la lista: {desv_estandar:.2f}")
        print(f"Máximo de la lista: {maximo:.2f}")
        
    except ValueError:
        # Histograma de frecuencias, por valores categóricos
        serie = pd.Series(lista).astype(str)
        histograma = serie.value_counts()
        
        for elemento, frecuencia in histograma.items():
            print(f"{elemento}: {frecuencia}")
```
