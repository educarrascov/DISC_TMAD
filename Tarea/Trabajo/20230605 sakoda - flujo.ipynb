## Generación Diagrama de Flujo SAKODA-SCHEELING:

Referencias: 
- [Graphwiz](https://graphviz.org/doc/info/shapes.html)
- [Diagramas](https://www.lucidchart.com/pages/es/que-es-un-diagrama-de-flujo#:~:text=Un%20diagrama%20de%20flujo%20es,claros%20y%20f%C3%A1ciles%20de%20comprender.)

```python
from graphviz import Digraph

# Crear una nueva gráfica dirigida
dot = Digraph('Modelo_Sakoda_Schelling')

# Agregar nodos al grafo
dot.node('A', 'Inicio', color='lightblue2', style='rounded,filled', shape='ellipse')
dot.node('B', 'Ingresar parámetros del modelo\n(n: tamaño grilla\n p: probabilidad de celda\n t: tolerancia\n pasos: número iteraciones)', 
         color='yellow', style='rounded,filled', shape='note', width='0.3', height='0.2')
dot.node('C', 'Inicializar grilla con probabilidades p \n(0: Vacía, 1: Azul, 2: Roja)', color='lightgreen', style='rounded,filled', shape='box')
dot.node('D', 'Iniciar Iteración:  \nSecuencial', color='lightgreen', style='rounded,filled', shape='box')
dot.node('E', 'Determinar totalidad celdas infelices en grilla actual\n(SI suma vecinos diferentes <= t, entonces feliz)', color='lightgreen', style='rounded,filled', shape='box')
dot.node('F', '¿Hay celdas infelices?', color='lightpink', style='rounded,filled', shape='diamond')
dot.node('G', 'Elegir celda infeliz al azar', color='lightgreen', style='rounded,filled', shape='box')
dot.node('H', '¿La vecindad de la celda infeliz\n tiene celdas vacías?', color='lightpink', style='rounded,filled', shape='diamond')
dot.node('I', 'Mover celda infeliz a una posición vacía en su vecindario\n (Elección aleatoria)', color='lightgreen', style='rounded,filled', shape='box')
dot.node('J', 'Actualizar grilla', color='lightgreen', style='rounded,filled', shape='box')
dot.node('K', '¿Hemos alcanzado el \nnúmero de iteraciones?', color='lightpink', style='rounded,filled', shape='diamond')
dot.node('L', 'Fin de una Iteración', color='lightgreen', style='rounded,filled', shape='box')
dot.node('M', 'Fin', color='lightblue2', style='rounded,filled', shape='ellipse')

# Agregar aristas entre los nodos
dot.edges(['AB', 'BC', 'CD', 'DE', 'EF'])
dot.edge('F', 'G', label='Si')
dot.edge('F', 'M', label='No')
dot.edge('G', 'H')
dot.edge('H', 'I', label='Si')
dot.edge('H', 'G', label='No', constraint='false')
dot.edges(['IJ', 'JK'])
dot.edge('K', 'M', label='Si')
dot.edge('K', 'L', label='No')
dot.edge('L', 'D')  # Conectar el fin de una iteración con el comienzo de la próxima

dot.render('modelo_sakoda_schelling_detailed_v20.gv', view=True)  # Guardar y abrir el diagrama de flujo
```
